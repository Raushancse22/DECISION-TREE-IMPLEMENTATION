```python
# Decision Tree Classifier - Iris Dataset
```

```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# 2: Load Dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

print("Feature Names:", feature_names)
print("Target Names:", target_names)
```

```python
# 3: Split Data into Training and Test Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

# Step 4: Build Decision Tree Model
dtree = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)
dtree.fit(X_train, y_train)
```

```python
# Step 5: Visualize the Decision Tree
class_names = list(iris.target_names)  # Ensuring this is a list
plt.figure(figsize=(15, 10))
plot_tree(dtree, feature_names=feature_names,
          class_names=class_names, filled=True)
plt.title("Decision Tree Visualization")
plt.show()
# Optional: Textual Representation
tree_rules = export_text(dtree, feature_names=feature_names)
print(tree_rules)
```

```python

# Step 6: Evaluate the Model
y_pred = dtree.predict(X_test)
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(
    y_test, y_pred, target_names=class_names))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Step 7: Additional Visualization - Feature Importance
importances = dtree.feature_importances_
plt.barh(feature_names, importances)
plt.title("Feature Importance")
plt.xlabel("Importance")
plt.show()
```
